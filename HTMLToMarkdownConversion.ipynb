### Saving a webpage as Markdown file

```python
import requests
import html2text
```

```python
def url_to_markdown(url):
    response = requests.get(url)
    response.raise_for_status()  # Raise error if request failed

    html_content = response.text
    markdown = html2text.html2text(html_content)

    return markdown
```

```python
outputPath = r"\\XGGC.SCOT.NHS.UK\GGCData\FolderRedirects\GRI\ELIASJO374\My Documents\review_of_Primary_Care_Guide\prescr_supply_admin_by_same_professional.txt"
urlAddress = "https://www.rpharms.com/recognition/all-our-campaigns/policy-a-z/prescribing-and-dispensing"
```

```python
def save_website_as_MD(urlAdress = urlAddress, outputPath = outputPath):
    with open(outputPath, "w", encoding="utf-8") as f:
        f.write(url_to_markdown(urlAddress))
```
